Defina una función que calcule la serie de Taylor del coseno hasta el n-ésimo término. Los inputs son n y el punto x donde se evalúa el coseno. Luego grafique el coseno y las aproximaciones con n=2 , n=6 en un rango de x que muestre dónde vale la aproximación y dónde se rompe.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

$$$$$$
\sum_{n=0}^{\infty}\frac{(-1)^n}{(2n)!}x^{2n}
$$$$$$

```python
def Taylor(n,x):
    '''
    Esta función evalua la serie de taylor del coseno utilizando los 
    argumentos de n que corresponde a la cantidad de términos de la sumatoria
    y x corresponde al valor del angulo que le damos al cos.
    '''
    cos=[1] #1 es el primer termino de la sumatoria de la expansion
    y=math.radians(x) #transformamos los numeros a radianes para poder trabajar más facil
    for i in range(1,n): #usamos un ciclo pa comenzar la sumatoria
        cos.append(cos[i-1] + ((-1)**i * y**(2*i))/(math.factorial(2*i))) #hacemos la sumatoria y la introducimos en una lista
    return cos # y nos entrega la lista
```

```python
#usando la función graficamos vs el valor real del coseno y obtenemos que mientras mayor sea el n
#más nos acercamos la valor real del coseno, por ejemplo en este caso con n=2 los valores estan
#bastante lejos del valor del coseno 

fig,ax=plt.subplots(1,4,figsize=(20,5))
ax[0].plot(np.arange(2),Taylor(2,45),label='Serie de Taylor del coseno',color='purple')
ax[0].plot(np.arange(2),np.full(2,np.cos(np.pi/4)),label='Valor real del coseno',color='red')
ax[0].legend()
 
ax[1].plot(np.arange(2),Taylor(2,90),label='Serie de Taylor del coseno',color='purple')
ax[1].plot(np.arange(2),np.full(2,np.cos(np.pi/2)),label='Valor real del coseno',color='red')
ax[1].legend()

ax[2].plot(np.arange(2),Taylor(2,180),label='Serie de Taylor del coseno',color='purple')
ax[2].plot(np.arange(2),np.full(2,np.cos(np.pi)),label='Valor real del coseno',color='red')
ax[2].legend()

ax[3].plot(np.arange(2),Taylor(2,270),label='Serie de Taylor del coseno',color='purple')
ax[3].plot(np.arange(2),np.full(2,np.cos(3*np.pi/2)),label='Valor real del coseno',color='red')
ax[3].legend()
```

```python
#si usamos un valor más grande de n podemos ver que el valor se acerca más al valor real
#pero si aumentamos el angulo que utilizamos en la función,la aproximación tiene dificultad
#para converger en el valor del cos real
fig,ax=plt.subplots(1,4,figsize=(20,5))
ax[0].plot(np.arange(6),Taylor(6,45),label='Serie de Taylor del coseno',color='purple')
ax[0].plot(np.arange(6),np.full(6,np.cos(np.pi/4)),label='Valor real del coseno',color='red')
ax[0].legend()

ax[1].plot(np.arange(6),Taylor(6,90),label='Serie de Taylor del coseno',color='purple')
ax[1].plot(np.arange(6),np.full(6,np.cos(np.pi/2)),label='Valor real del coseno',color='red')
ax[1].legend()

ax[2].plot(np.arange(6),Taylor(6,180),label='Serie de Taylor del coseno',color='purple')
ax[2].plot(np.arange(6),np.full(6,np.cos(np.pi)),label='Valor real del coseno',color='red')
ax[2].legend()

ax[3].plot(np.arange(6),Taylor(6,270),label='Serie de Taylor del coseno',color='purple')
ax[3].plot(np.arange(6),np.full(6,np.cos(3*np.pi/2)),label='Valor real del coseno',color='red')
ax[3].legend()
```
