```python
import matplotlib.pyplot as plt #Importamos para graficar
import numpy as np            #Importamos para trabajar con numpy
import math                  #Importamos para usar los factoriales

```

Para calcular $\pi$ utilizamos la serie de Taylor del coseno inverso de 1/2 la cual es:
$$
\arccos(\frac{1}{2})=\frac{\pi}{2}-\arcsin(\frac{1}{2})
$$
Si expandimos el seno inverso nos queda y resolvemos el coseno inverso:
$$
\frac{\pi}{3}-\frac{\pi}{2}=\sum_{n=0}^{\infty}\frac{(2n)!}{4^n(n!)^2(2n+1)}x^{2n+1}
$$
y finalmente queda:
$$
\pi=6\sum_{n=0}^{\infty}\frac{(2n)!}{4^n(n!)^2(2n+1)}x^{2n+1}
$$

```python
piclase1 = [4] #Estas son las utilizadas en clase que son Leibniz
piclase2 = [3] #y esta es tangente inversa en Taylor
pi1=[3]        #si usamos el despeje que se hizo al principio del cuadernillo podemos escribir la formula de pi como la siguiente
       #el primer termino es 3 debido al arccos(1/2)
for n in range(1, 100):
    piclase1.append(piclase1[n-1] + (-1)**n*4/(2*n + 1))           #utilizamos el append para crear un conjunto que corresponda a lo que queremos
    piclase2.append(piclase2[n-1] + (-1)**(n + 1)*4/(2*n*(2*n + 1)*2*(n+1)))
    pi1.append(pi1[n-1]+6*(math.factorial(2*n)/((4**n)*((math.factorial(n))**2)*(2*n+1)))*(x)**(2*n+1)) 
 #utilizamos el append para crear un conjunto que corresponda a lo que queremos y poder graficar
```

```python
pi2=[]                      #Esta es una manera de hacerlo donde despejo pi desde el resultado de arccos(1/2) pero utilizo pi
                            #en el proceso entonces no es el mejor
x=0.5                    
for n in range(0,100):
    pi2.append(2*((np.pi/2)-(math.factorial(2*n)/((4**n)*((math.factorial(n))**2)*(2*n+1)))*(x)**(2*n+1)))
```

```python
pi3=[]                    #Esta es otra manera donde se despeja el pi/2 dentro de la expansion en taylor del arccos 
                         #pero termina dando pequeños problemas numericamente y hay que ajustarlo.
x=0.5
for n in range(0,100):
    pi3.append(3*(np.arccos(1/2)+(math.factorial(2*n)/((4**n)*((math.factorial(n))**2)*(2*n+1)))*(x)**(2*n+1)))
```

```python
fig,ax = plt.subplots(1,3,figsize=(20,5))
ax[0].loglog(np.arange(100), pi1 , label= '$\pi$ calculado',color="purple")
ax[0].loglog(np.arange(100), piclase1 , label= 'Leibniz',color="yellow")
ax[0].loglog(np.arange(100), piclase2 , label= 'Taylor',color="orange")
ax[0].loglog(np.arange(100), np.full(100, np.pi), label = '$\pi$ real',color="red") 
ax[0].legend() #graficamos cada uno de los pi que obtuvimos en diferentes figuras comparando con las series de la clase
ax[0].set_xlim(1,30) #escogi este numero porque es donde se puede ver mejor la diferencia entre cada caso
ax[0].set_ylim(np.pi - 0.001*np.pi, np.pi + 0.001*np.pi)

ax[1].loglog(np.arange(100), pi2 , label= '$\pi$ 2 calculado',color="green")
ax[1].loglog(np.arange(100), piclase1 , label= 'Leibniz',color="yellow")
ax[1].loglog(np.arange(100), piclase2 , label= 'Taylor',color="orange")
ax[1].loglog(np.arange(100), np.full(100, np.pi), label = '$\pi$ real',color="red") 
ax[1].legend()
ax[1].set_xlim(1,30)
ax[1].set_ylim(np.pi - 0.001*np.pi, np.pi + 0.001*np.pi)

ax[2].loglog(np.arange(100), pi3 , label='$\pi$ 3 calculado',color="blue")
ax[2].loglog(np.arange(100), piclase1 , label= 'Leibniz',color="yellow")
ax[2].loglog(np.arange(100), piclase2 , label= 'Taylor',color="orange")
ax[2].loglog(np.arange(100), np.full(100, np.pi), label = '$\pi$ real',color="red") 
ax[2].legend()
ax[2].set_xlim(1,30) 
ax[2].set_ylim(np.pi - 0.001*np.pi, np.pi + 0.001*np.pi)
```

Finalmente podemos ver que la serie de taylor del coseno inverso converge mucho mas rapidamente que las obtenidas en clase
